```python
df1 = df[df["Price"] < 30000]
df2 = df[(df["Price"] >= 30000) & (df["Price"] <= 60000)]
df3 = df[df["Price"] > 60000]
with pd.ExcelWriter('Prices.xlsx') as writer:  
    df1.to_excel(writer, sheet_name='LessThan30000')
    df2.to_excel(writer, sheet_name='Between30000And60000')
    df3.to_excel(writer, sheet_name='GreaterThan60000')
```
